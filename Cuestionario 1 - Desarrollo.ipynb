Pregunta 1



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def r(x): #Moto
    return 1.4*(x**2) - 5.6*x + 12.6

def s(x): #Auto
    return -1.6*(x**2) + 1.3*x + 141.3

def inter(x): #Interseccion
    return r(x)-s(x)

x = np.linspace(1, 10, 4)
sol = np.around(fsolve(inter, x), 1)
interseccion = np.unique(sol)

print(f'El(los) punto(s) de interseccin de las funciones est(n) en x = {interseccion}')
print(f"s(7.8) = {s(7.8):.1f}")
print(f"r(7.8) = {r(7.8):.1f}")


plt.plot(x, s(x),color = 'blue', label = 'Auto')
plt.plot(x, r(x),color = 'orange', label = 'Moto')
plt.title('Velocidad pos Robo')
plt.ylabel('Kilometros por hora')
plt.xlabel('Minutos')
plt.ylim(0,)
plt.grid(True)
plt.legend()
plt.show()
```

Pregunta 15



```python
import numpy as np
from scipy.optimize import fsolve

def I(x):
    return 50*x + 0.5*(x**2)

def C(x):
    return 10*x + 450

def inter(x):
    return I(x)-C(x)

x = np.linspace(1, 50, 10)
solucion = np.around(fsolve(inter, x),1)

# np.unique() permite eliminar las soluciones repetidas
interseccion = np.unique(solucion)

print(f'El(los) punto(s) de interseccin de las funciones est(n) en x = {interseccion}')

plt.plot(x, I(x),color = 'blue', label = 'nombre de la funcin f')
plt.plot(x, C(x),color = 'orange', label = 'nombre de la funcin g')
plt.title('Nombre del grfico')
plt.ylabel('Nombre eje y')
plt.xlabel('Nombre eje x')
plt.grid(True)
plt.legend()
plt.show()
```
